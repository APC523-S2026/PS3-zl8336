This system is undergoing a forced vibration process. The fast vibration is under the system's natural frequency. Then there is a slow modification under the external force's frequency.

(a) forward Euler.

The energy is shifting.

We can see the time cost for Forward Euler method will be

24 ms ± 172 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)

```python
#%%timeit
p2.solve(
    t_stop=100,
    N=1000000,
    u0=jnp.array([0.0, 0.0],dtype=jnp.float64),
    method_name='Forward_Euler',
    plot=True,
    )
```

(b) symplectic Euler.

The energy is correct. The required running time is close to the time in forward Euler method.

We can see the time cost for Symplectic Euler method will be

32 ms ± 208 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)

The time cost of Symplectic Euler method is close to that of Forward Euler method. 
Symplectic Euler method has better in energy conservation.

```python
#%%timeit
p2.solve(
    t_stop=100,
    N=1000000,
    u0=jnp.array([0.0, 0.0],dtype=jnp.float64),
    method_name='Symplectic_Euler',
    plot=True,
    )
```

(c) RK4

The energy is correct.

We can see the time cost for RK4 method will be

57.1 ms ± 180 μs per loop (mean ± std. dev. of 7 runs, 10 loops each)

It uses about twice of the time.

It is the most accurate method.

```python
#%%timeit
p2.solve(
    t_stop=100,
    N=1000000,
    u0=jnp.array([0.0, 0.0],dtype=jnp.float64),
    method_name='RK4',
    plot=True,
    )
```
